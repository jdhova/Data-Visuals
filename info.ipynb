```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
data = pd.read_csv('./Data/housing.csv')
data
```

```python
data2 = data['ocean_proximity'].value_counts().rename_axis('sub_values').reset_index(name='counts')
data2
```

```python
data3 = data2.drop(4)
data3
```

```python
labels = data3.sub_values
values = data3.counts
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')


ax.pie(values,labels = labels, autopct = '%1.2f%%')
plt.show()
```
